## Determine the five most common journals and the total articles for each

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
file = '2 file.csv'
df = pd.read_csv(file, encoding='iso-8859-1')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
new_list =[]
for row in df['Journal title']:
    separated_row = row.split(' ')
    row = [word.lower() for word in separated_row]
    row = ' '.join(row)
    new_list.append(row)

df['Journal title 2'] = new_list
common_journals = df.groupby('Journal title 2')['Journal title 2'].count().sort_values(ascending=False)[:5]
print(common_journals)
```
