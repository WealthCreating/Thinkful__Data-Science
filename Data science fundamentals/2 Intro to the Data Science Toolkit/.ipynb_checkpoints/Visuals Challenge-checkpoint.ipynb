```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = 'titanic.csv'
df = pd.read_csv(file)
list(df)
```

```python
plt.scatter(x=df['Age'], y=df['Fare'])
plt.ylabel('Ticket Cost in Dollars')
plt.xlabel('Passenger Age')
plt.title('Ticket Price Compared to Passenger Age')
plt.show()
```

```python
df[df['Fare'] > 500]['Name']
```

```python
fig1 = df[df.Sex=="male"]
fig2 = df[df.Sex=="female"]

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.scatter(x=fig1['Age'],y=fig1['Fare'], color='green')
plt.title("Male")

plt.subplot(1, 2, 2)
plt.scatter(x=fig2['Age'],y=fig2['Fare'], color='blue')
plt.title("Female")


plt.show()
```

```python
group = df.groupby(['Sex'])['Sex'].count()
plt.hist(group)
plt.show()
```
